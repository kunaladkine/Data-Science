```python
a=np.random.randint(10,30,size=15)
print("content of a")
print(a)
```

```python
#find the max and min value fromm 1 dim
maxv=np.amax(a)
minv=np.amin(a)
print("max value =",maxv)
print("min value=",minv)
```

```python
#find max and min value from 2d dim
a=np.random.randint(10,30,size=(4,4))
print(a)
maxv=np.amax(a)
minv=np.amin(a)
print("max value =",maxv)
print("min value=",minv)
```

```python
#Now My Req: To find Column-wise and Row-wise Max and Min Values
#To Do the above, we must use axis concept.
# if axis=0 then It refers COLUMNS
# if axis=1 then It refers ROWS
maxv=np.amax(a)
minv=np.amin(a)
colmax=np.amax(a,axis=0)
rowmax=np.amax(a,axis=1)
colmin=np.amin(a,axis=0)
rowmin=np.amin(a,axis=1)
print("--------------------------------------")
print("Given Matrix")
print(a)
print("--------------------------------------")
print("Max Value=",maxv)
print("Min Value=",minv)
print("--------------------------------------")
print("Column Max=",colmax)
print("Row Max=",rowmax)
print("Column Min=",colmin)
print("Row Min=",rowmin)
print("--------------------------------------")
```

```python
#Find the mean on ndarray by using mean()--1-Dim
a=np.array([3,1,2,4])
m=np.mean(a)
print("Mean({})={}".format(a,m))
```

```python
#Our Own Code for Finding Sum and Mean (Average)
s=0
for v in a:
    s=s+v
print("Sum=",s)
print("mean({})={}".format(a,s/len(a)))
```

```python
#Find the mean on ndarray by using mean()--2-Dim
#Find the mean for Col and Row
a=np.array([3,1,2,4]).reshape(2,2)
print("-------------------------------------")
print("Matrix-A")
print(a)
print("-------------------------------------")
m=np.mean(a)
cm=np.mean(a,axis=0)
rm=np.mean(a,axis=1)
print("Mean\n({})={}".format(a,m))
print("Col Mean\n({})={}".format(a,cm))
print("Row Mean\n({})={}".format(a,rm))
print("-------------------------------------")
```

```python
#Finding Variance by using var()
a=np.array([3,1,2,4])
print("Content of a")
print(a)
v=np.var(a)
print("Variance=",v)
```

```python
#Find Col and Row-wise variance
a=np.array([3,1,2,4]).reshape(2,2)
print("------------------------------------")
print("Content of a")
print(a)
print("------------------------------------")
v=np.var(a)
print("Variance=",v)
cv=np.var(a,axis=0)
rv=np.var(a,axis=1)
print("Col Var=",cv)
print("Row Var=",rv)
print("------------------------------------")
```

```python
#Find Col and Row-wise variance
a=np.arange(4*3).reshape(4,3)
print("------------------------------------")
print("Content of a")
print(a)
print("------------------------------------")
v=np.var(a)
print("Variance=%0.2f" %v)
cv=np.var(a,axis=0)
rv=np.var(a,axis=1)
print("Col-wise Var")
for v in cv:
    print(round(v,2),end="  ")
print()    
print("Row-wise Var")
for v in rv:
    print(round(v,2),end="  ")
print()    
print("------------------------------------")
```

```python
#We Know that standard Deviation= sqrt(Var)
#Find Col and Row-wise variance
a=np.array([3,1,2,4]).reshape(2,2)
print("------------------------------------")
print("Content of a")
print(a)
print("------------------------------------")
s=np.std(a)
print("Standard Deviation=",s)
csd=np.std(a,axis=0)
rsd=np.std(a,axis=1)
print("Col Std =",csd)
print("Row Std=",rsd)
print("------------------------------------")
```
