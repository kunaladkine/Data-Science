```python
import numpy as np
a=np.array([10,20,30,40,50,60,70,80])
print(a,type(a))
print("Shape of Array=",a.shape)
print("Dim of Array=",a.ndim)
print("Size of Array=",a.size)
```

```python
#to get the value 10
print(a[0])
```

```python
#to get the value of 70
print(a[-2])
```

```python
#to get the value of 50
print(a[4])
```
