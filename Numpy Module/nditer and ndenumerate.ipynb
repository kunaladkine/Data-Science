```python
#Iterating Array Without Fun
import numpy as np
from numpy import random
```

```python
arr=np.array([10,20,30,40,50,60])
for x in arr:
    print(x)
```

```python
arr2=np.array([[1,2,3,4,5],[10,20,30,40,50]])
for x in arr2:
    print(arr2)
```

```python
for x in arr2:
    for y in x:
        print(y)
```

```python
arr=np.array([[[1,2,3],[4,5,6]],[[12,13,14],[16,17,18]]])
for x in arr:
    print(x)
```

```python
for x in np.nditer(arr):
    print(x)
```

```python
for idx,x in np.ndenumerate(arr):
    print(idx,"--->",x)
```

```python
import numpy as np
from numpy import random as r
arr2=r.randint(1,12,size=12)
print(arr2)
```

```python
for x in np.nditer(arr2):
    print(x)
```
