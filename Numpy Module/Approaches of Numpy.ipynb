```python
#Array 

import numpy as np
lst=[10,20,30,40,50,60]
print("Origrinal List =",lst)
lst1=np.array(lst)
print("After Converted List =",lst1)
print("----------------------")
print("shape of Array=",lst1.shape)
print("Size of Array =",lst1.size)
print("Dim of Array =",lst1.ndim)
print("Reshape of Array =")
print(lst1.reshape(3,2))
```

```python
#arange

#varname=numpy.aragne(value)

import numpy as np
a=np.arange(10)
print(a)
print("----------------------")
print("shape of Array=",a.shape)
print("Size of Array =",a.size)
print("Dim of Array =",a.ndim)
```

```python
#hstack

#varname=numpy.hstack((ndarrayobj1,ndarrayobj2))

a=np.array([[10,20,30,40],[50,60,70,80]])
print("array1")
print(a)
b=np.array([[1,2,3,4],[3,4,4,5]])
print(b)
```

```python
c=np.hstack((a,b))
print("Merged Array")
print(c)
```

```python
c=np.vstack((a,b))
print("Merged Array")
print(c)
```

```python
c=np.hstack((a,b))
d=c.reshape(3,2)
print(d)
```
