```python
import numpy as np
from numpy import random as r
a=np.array([10,20,30,40,50,60])
print(a,type(a))
```

```python
b=a.reshape(3,2)
print(b)
```

```python
#shallow copy is = copy
c=a.copy()
print(a,type(a),id(a))
print("-----------------------")
print(c,type(c),id(c))
```

```python
#deep copy is = view
c=a
print(a,id(a))
print(c,id(c))
```

```python
a[0]=100
print(a)
print("The id of a is =",id(a))
```

```python
print(c)
print("The id of c=",id(c))
```

```python
#view
arr=np.array([10,20,30,40,50,60,70,80])
arr2=arr.view()
print(arr2,id(arr2),type(arr2))
```

```python
print(arr,id(arr),type(arr))
```

```python
print("The Base of arr=",arr.base)
print("The Base of arr2=",arr2.base)
```
