```python
import numpy as np
from numpy import random as r
a=r.randint(0,100,size=(16)).reshape(4,4)
print(a)
```

```python
#to get the value of all ends
a[::3,::3]
```

```python
#to get the values of last two col
a[::,2:4]
```

```python
#To get the value of 51,26
a[1:2,2:4]
```

```python
#To Get the Value of 41,3
a[2:3,1:3]
```

```python
a[-1:,-4:-3]
```

```python
a[0:,0:2]
```

```python
a[2:,0::]
```

```python
a[0:1,3:,0:1]
```

```python
a[0:1,2:3,1:]
```
