```python
import numpy as np
from numpy import random as r
a=np.array([10,20,30,40,50,60,70,80,90,12,55,66]).reshape(2,3,2)
print(a)
```

```python
#converting 3darray to 1darray
c=a.flatten()
print(c)
```

```python
a=np.array([10,20,30,40,50,60,70,80,90,12,55,66]).reshape(2,3,2)
print(a)
d=a.ravel()
print(d)
```

```python
#converting 2d array to 1d array
a=np.array([10,20,30,40,50,60]).reshape(3,2)
print(a)
```

```python
b=a.flatten()
print("Flatten Array =",b)
```

```python
#to convert the 2d,3rd, nd array to 1d array we use the flatten and ravel.
```
