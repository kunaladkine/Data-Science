```python
#deleting the rows & columns from the 2D numpy Array
arr=np.array([[11,22,11,22],
             [22,33,23,23],
             [43,43,54,21],
             [66,45,67,23]])
print(arr)
```

```python
#column will be deleted at 1th index
arr=np.delete(arr,1,axis=1)
print(arr)
```

```python
#row will be deleted at 0 th index
arr=np.delete(arr,0,axis=0)
print(arr)
```

```python
#adding row to the 0th index
arr=np.insert(arr,0,[22,22,22],axis=0)
print(arr)
```

```python
#adding column to the -1 position
arr=np.insert(arr,-1,[[12,22,33,44]],axis=1)
print(arr)
```

```python
#deletig the column 3 to -3
arr=np.delete(arr,[3,4],axis=1)
print(arr)
```
