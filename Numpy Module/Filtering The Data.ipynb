```python
#get multiple 3 using the numpy.where()
a=r.randint(1,50,size=20)
ind=np.where(a%3==0)
print(ind)
for i in ind[0]:
    print(i)
```

```python
#to get the +ve elements by using numpy.where()
ind=np.where(a>0)
print(ind)
for i in ind[0]:
    print(a[i])
```

```python
#working with missing data
import numpy as np
data=np.array([1,2,None,3,4])
print(data)
```

```python
#Replace misssing with 0
data=np.where(data==None,0,data)
print(data)
```

```python
#filter missing values
data=np.extract(data!=None,data)
print(data)
```

```python
#Example on numpy.extract(TestCond,ndarrayobj)
import numpy as np
from numpy import random
a=np.random.randint(2,40,size=14)
print(a)
#get the value whiech are more than 20
print("-----------------------")
morethan20=np.extract(a>20,a)
print("Greater Than 20=",morethan20)
```
