```python
import numpy as np
a=np.array([10,20,30,40,50,60,70,80,90,15,25,35])
print("-------------------------------------------------")
print("Content of a")
print(a)
print("Dimension of a=",a.ndim)
print("Shape of a=",a.shape)
print("Size of a=",a.size)
print("-------------------------------------------------")
```

```python
import numpy as np
a=np.array([10,20,30,40,50,60,70,80,90,15,25,35]).reshape(3,4)
print("-------------------------------------------------")
print("Content of a")
print(a)
print("Dimension of a=",a.ndim)
print("Shape of a=",a.shape)
print("Size of a=",a.size)
print("-------------------------------------------------")
```
