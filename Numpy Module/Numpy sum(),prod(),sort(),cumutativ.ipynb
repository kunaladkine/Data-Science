```python
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([2,4,6])
print(arr1)
print(arr2)
print("----------------------")
#adding the elements 
result=np.add(arr1,arr2) #1+1,2+4, 3+6
print("Addition of Array",result) 
```

```python
#using the Sum
arr1=np.array([1,2,3])
arr2=np.array([2,4,6])
print(arr1)
print(arr2)
print("----------------------")
#adding the elements 
result=np.sum([arr1,arr2]) 
print("Addition of Array",result) 
```

```python
#adding the element ussing axix 1 sum
result=np.sum([arr1,arr2],axis=1)
print(result)
```

```python
arr=np.array([1,2,3,4])
print("original array",arr)
result=np.cumsum(arr)
print("Cumutative sum",result)
```

```python
arr=np.array([3,2,1,3,5])
print("original array",arr)
result=np.cumsum(arr)
print("Cumutative sum",result)
```

```python
import numpy as np
arr = np.array([1, 2, 3, 4])
x = np.prod(arr)
print(x) #  Returns: 24 because 1*2*3*4 = 24
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
x = np.prod([arr1, arr2])
print(x) # Returns: 40320 because 1*2*3*4*5*6*7*8 = 40320
```

```python
arr = np.array([[1,2,3],[3,4,5]])
x = np.prod(arr,axis=0) # Here axis=0 Represents Column Product
print(x)
```

```python
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])
newarr = np.prod([arr1, arr2], axis=1) # Here axis=1 Represents Row Product
print(newarr)# Returns: [24   1680]
```

```python
arr1=np.array([1,2,4,5])
arr2=np.array([9,5,3,2])
prdarr=np.prod([arr1,arr2],axis=0)
print(prdarr)
```

```python
arr=np.array([1,2,3,4])
newarr=np.cumprod(arr)
print(newarr)
```

```python
arr = np.array([[1,2,3],[3,4,5],[2,3,5]])
print("Content of arr")
print(arr)
print("-------------------------")
x = np.cumprod(arr,axis=1) # Here axis=1 Represents Cumulative Row Product 
print(x)
```

```python
arr = np.array([[1,2,3],[3,4,5],[2,3,5]])
print("Content of arr")
print(arr)
print("-------------------------")
x = np.cumprod(arr,axis=0) # Here axis=0 Represents Cumulative Column Product
print(x)
```

```python
arr=np.array([3,2,1,3,5])
print("original array",arr)
result=np.sort(arr)
print("sorted Array",result)
```

```python
arr=np.array([[3,2,1],[4,3,5]])
print(arr)
newarr=np.sort(arr)
print(newarr)
```

```python
arr=np.array([[3,2,1],[4,3,5]])
print(arr)
newarr=np.sort(arr,axis=0) #col wise
print(newarr)
```

```python
arr=np.array([[3,2,1],[4,3,5]])
print(arr)
newarr=np.sort(arr,axis=1) #row wise
print(newarr)
```
