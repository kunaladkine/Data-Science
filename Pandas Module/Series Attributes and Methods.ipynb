```python
import numpy as np
import pandas as pd
lst=["Rossum","Travis","Ritche","Dennis","Hunter"]
s=pd.Series(lst)
print("Series Object Data")
print(s)
```

```python
lst=["Rossum","Travis","Ritche","Dennis","Hunter"]
s=pd.Series(lst,index=["a","b","c","d","e"])
print("Series Object Data")
print(s)
```

```python
s["a"]
```

```python
lst=["Rossum","Travis","Ritche","Dennis","Hunter"]
s=pd.Series(lst,index=["a","b","c","d","e"],name="student")
print("Series Object Data")
print(s)
```
