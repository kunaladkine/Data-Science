```python
c=pd.Series([10,33,24,55,22,34,91,0,3],index=["a","b","c","d","e","f","g","h","i"])
print(c)
```

```python
#Sorting the Values of Series
#Syntax:   seriesobj.sort_values(ascending,inplace)
#seriesobj.sort_values()
#seriesobj.sort_values(inplace)
#seriesobj.sort_values(ascending)
```

```python
print(c.sort_values())
```

```python
print(c.sort_values(ascending=True))
```

```python
print(c.sort_values(ascending=False))
```

```python
print(c.sort_values(inplace=False))
```

```python
print(c.sort_values(ascending=True,inplace=False))
```

```python
#to print the ascending values using for loop
for val in c.sort_values():
    print(val)
```

```python
#To Print the Descending Value using for loop
for val in c.sort_values(ascending=False):
    print(val)
```

```python
#Finding Max / Highest Element from Series and Number Approaches
#SeriesObj.max()
#SeriesObj.nlargest(n)
```

```python
c=pd.Series([10,20,30,40,50,3,66,99,32,45,94])
print("Series is =\n",c)
```

```python
#to find the max no of element form series
print(c.max())
```

```python
print("Max no is =",c.max())
```

```python
#using the nlargest(n)
print(c.nlargest())
```

```python
print(c.nlargest(1))
```

```python
#to get no of largest element
print(c.nlargest(3))
```

```python
#to get the sum of first 2 largest no
print(c.nlargest(2).sum())
```

```python
#Finding Min / smallest Element from Series and Number Approaches
#SeriesObj.min()
#SeriesObj.nsmallest(n)
```

```python
c=pd.Series([10,20,30,40,50,3,66,99,32,45,94])
print("Series is =\n",c)
```

```python
#to find the minimum no of element from the series
print(c.min())
```

```python
print("The Minimum no of Series is =",c.min())
```

```python
#to print the smallest no using nsmallest(n)
print("Smallest No =\n",c.nsmallest())
```

```python
#to print the 1st three smallest no
print(c.nsmallest(3))
```

```python
#to print the sum of 1sst three smallest no 
print(c.nsmallest(3).sum())
```
