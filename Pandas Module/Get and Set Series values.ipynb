```python
#Access multiple elements of Series by specific index positions
names = pd.Series(  ['Mark', 'Rita', 'Vicki', 'Justin', 'John', 'Michal'],
                    index = ['a', 'b', 'c', 'd', 'e', 'f'])
# Select elements at index position 2, 3 and 0 only
few_names = names[[2, 3, 0]]
# Display the subset of Series
print(few_names)
```

```python
print(names[[3,4]])
```

```python
few_names = names[["a","b","c"]]
# Display the subset of Series
print(few_names)
```

```python
few_names = names[["c","d"]]
# Display the subset of Series
print(few_names)
```

```python
#Changing elements in the Series
names = pd.Series(  ['Mark', 'Rita', 'Vicki', 'Justin', 'John', 'Michal'],index = ['a', 'b', 'c', 'd', 'e', 'f'])
print(names)
```

```python
names[0]="Ravi"
print(names[0])
```

```python
print(names)
```

```python
names[[0,1]]="Teja","Praks"
print(names)
```

```python
names["e"]="Stone"
print(names)
```

```python
names[["c","e"]]="Don","Son"
print(names)
```

```python
#Change multiple elements in Series
names[1:3]="Ravi Teja"
print(names)
```

```python
names[0:2]="Ram Charan"
print(names)
```

```python
names["a":"d"]="Allu Arjun"
print(names)
```

```python
names["a":"b"]="MB"
print(names[["a","b"]])
```
