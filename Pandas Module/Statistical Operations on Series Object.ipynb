```python
#Statistical Operations on Series Object
import numpy as np
import pandas as pd
a=pd.Series([10,20,30,40,50])
print("Series is \n",a)
```

```python
#seriesobj.sum()
#to find the sum of element in series
print(a.sum())
```

```python
print("Sum of Numbers is =",a.sum())
```

```python
#seriesobj.count()
#to print the no of elements in series
print(a.count())
```

```python
#seriesobj.max()
#to print the maximum no of element 

print(a.max())
```

```python
#seriesobj.min()
#to print the minimum no of element

print(a.min())
```

```python
#seriesobj.mean()
#to print them mean of series

print(a.mean())
```

```python
#seriesobj.median()
#to print the median of series

print(a.median())
```

```python
#serieobj.var()

print(a.var())
```

```python
#seriesobj.std()

print(a.std())
```

```python
#seriesobj.mode()

print(a.mode())
```

```python
# seriesobj.describe()
#to get all the valuess from count, mean, std, min, max 
print(a.describe())
```

```python
#print(seriesobj.agg(["max","min","mean".........]))

# to print the min , max ,mean aggreagation at a time 

print(a.agg(["max","min","mean","std"]))
```

```python
#seriesobj.product()

print(a.product())
```

```python
#seriesobj.cumprod()

print(a.cumprod())
```
